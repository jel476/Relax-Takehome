```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
```

```python
engagement = pd.read_csv('takehome_user_engagement.csv')
engagement['time_stamp'] = pd.to_datetime(engagement['time_stamp'], format= '%Y/%m/%d %H:%M:%S')
engagement
```

```python
engagement['wk'] = engagement['time_stamp'].dt.week
engagement = engagement.groupby(['wk','user_id'])['time_stamp'].count().reset_index(name="freq")

engagement
```

```python
result = engagement[engagement.freq >= 3]
final = result.groupby('user_id').count()[['freq']]
final
```
